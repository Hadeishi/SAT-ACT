##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
columns = sat_act_df.columns[1:]
```

```python
def std_dev_for_column(column):
    mean=np.mean(column)
    return (sum([(num - mean) ** 2 for num in column])/ len(sat_act_df)) ** .5
```

```python
sd = [std_dev_for_column(sat_act_df[column]) for column in columns]
sd
```

```python
sat_act_df.describe().T   
```

The values we obtained differ from those given by python since python assumes that the standard deviations we want are of a sample, not the entire population. Since we are interested in the standard deviation of our entire population, we divided our summed variances by n, as opposed to n-1.

##### 19. When we make assumptions about how data are distributed, what is the most common assumption?

In my experience, the most common assumption made by statisticians about how data are distributed is that they are normally distributed.

##### 20. Does this assumption hold true for any of our columns? Which?

Since normal distributions generally refer to the distribution of data gathered in comparable ways tending to occur around a mean in a symmetrical fashion, thus forming the canonical "bell shaped curve," if our histograms were to appear anything like a bell shaped curve it would be interesting but uninformative since the data that is represented by each block in the above histograms each represents something completely different from the others.  In other words, one block might represent the average of 200,000 student scores representing 98% of the graduating seniors of one state while another may represent the average of just 200 of another state's most eager and certainly college-bound seniors.

It is therefore unsurprising that looking at histograms of our columns above or in the pairplot below there is little evidence of normal distributions in our dataset. In fact, there may be some indication of a bimodal distribution in some of the variables, particularly in the SAT scores. Moreover, there is very strong skew in our two measures of participation, albeit in opposite directions:  there were a large number of states that showed 100% participation in the ACT while at the same time there were a large number of states that had very low participation in the SAT.

It is certainly true that were our samples gathered and displayed in a more typical fashion, say as an actual sample of test scores gathered from a single population, say graduating seniors from the state of California, I am confident that we would very likely see a normal distribution. Both SAT and IQ tests have been shown to be good predictors of "general intelligence," or g (Frey and Detterman, 2003 <https://www.psychologicalscience.org/pdf/ps/Frey.pdf?origin=publication_detail>), and g is very well-known to be normally distributed in populations.

Seeing that there was no reason to doubt the results of a t-test comparison of means even given substantial skew in the variables I am comparing, I decided to go forward with the test.

##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

So, to formalize the hypothesis test aluded to above, there are two variables, SAT participation and ACT participation whose mean values I believe are significantly different from one another. Understanding that science never actually *proves* anything per se--- in fact, to say that something has been proven true and is now "in the scientific canon" is itself inherently anti-scientific as it implies that these truths can never be disproven. In contrast, science is only able to show that things are so unlikely to happen that it is, for all intents and purposes, not true.

So, since the scientific method can only disprove things that are not true, in order to be truly useful, we set up experiments to test "null hypotheses" that are actually the opposite of what we want to prove. In our case,

my H0:  There is no difference between participation_sat and participation_act, the rates at which students in the various states participate in the college entrance exams.

my H1:  There is a difference between participation_sat and participation_act.

My alpha = .05, which means 5% of the time I will mistakenly reject the null hypothesis. In other words, I am accepting that 1 out of 20 times I run an experiment where alpha = .05 I will say there is a difference between my two variables when there isn't a difference.

```python
SAT_participation = np.array(sat_act_df['Participation_sat'].values)
ACT_participation = np.array(sat_act_df['Participation_act'].values)

print(np.mean(SAT_participation))
print(np.mean(ACT_participation))
```

```python
import scipy.stats as stats
```

```python
t_stat, p_value = stats.ttest_ind(ACT_participation, SAT_participation)
t_stat, p_value
```

p < .05 (actually, p = 0.00024, highly significant), so the null hypothesis is rejected.
Participation is significantly different in states' participation in the ACT vs. the SAT.
More states prefer the ACT; that was a surprise to me...

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

To generate 95% confidence intervals for SAT and ACT participation rates, I will assume a normal distribution (even though I know this is ***very*** far from the case. I don't know how to address this question otherwise, so here goes...)

```python
#  z = 1.95  # for 95%
#  ci_lower = single_mean-z*single_sem
#  ci_upper = single_mean+z*single_sem
```

```python
print(np.mean(SAT_participation))
print(np.mean(ACT_participation))
```

```python
import scipy.stats as stats
```

```python
SAT_95_conf_int = stats.sem(sat_act_df['Participation_sat']) * 1.95
```

```python
print(np.mean(SAT_participation) - SAT_95_conf_int)
```

```python
print(np.mean(SAT_participation) + SAT_95_conf_int)
```

```python
ACT_95_conf_int = stats.sem(sat_act_df['Participation_act']) * 1.95
```

```python
print(np.mean(ACT_participation) - ACT_95_conf_int)
```

```python
print(np.mean(ACT_participation) + ACT_95_conf_int)
```

So, For SAT participation, 95% of the time you will randomly pick a value between .302 and .494 and for ACT participation, 95% of the time you will randomly pick a value between .545 and .740. These confidence limits are clearly not overlapping.
